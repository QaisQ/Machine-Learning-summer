```python
import csv

with open('student-mat.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';') #csv_reader is a variable name

    with open('new-student-mat.csv', 'w') as new_file:
        csv_writer = csv.writer(new_file, delimiter='\t')
    
        for line in csv_reader:
            csv_writer.writerow(line)
   
```
